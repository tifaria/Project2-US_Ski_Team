```python
import time
from splinter import Browser
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
# Create path and setup local chrome browser
executable_path = {'executable_path': 'chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Alpine Team Page
url = 'https://usskiandsnowboard.org/alpine'
browser.visit(url)
```

```python
# Pull HTML and Beautiful Soup
html = browser.html
soup = bs(html, 'html.parser')
```

```python
# Alpine Name Loop
alpine_name = []
for items in soup.find_all(class_="views-field views-field-field-img-head-shot"):
    a_name = items.find_next_sibling().text
    alpine_name.append(a_name)
```

```python
# Alpine Hometown Loop
alpine_hometown = []
for items in soup.find_all(class_="views-field views-field-title"):
    a_hometown = items.find_next_sibling().text
    alpine_hometown.append(a_hometown)
```

```python
# Create all lists into DataFrames
df1 = pd.DataFrame(alpine_name) 
df2 = pd.DataFrame(alpine_hometown)
df3 = pd.DataFrame(freeski_name)
df4 = pd.DataFrame(freeski_hometown)
df5 = pd.DataFrame(freestyle_name)
df6 = pd.DataFrame(freestyle_hometown)
df7 = pd.DataFrame(xc_name)
df8 = pd.DataFrame(xc_hometown)
df9 = pd.DataFrame(ski_jumping_name)
df10 = pd.DataFrame(ski_jumping_hometown)
df11 = pd.DataFrame(nordic_combined_name)
df12 = pd.DataFrame(nordic_combined_hometown)
df13 = pd.DataFrame(snowboard_name)
df14 = pd.DataFrame(snowboard_hometown)
```

```python
# Rename each column depeding on whether it is a name or hometown
alpine_name_df = df1.rename(columns={ df1.columns[0]: "Name" })
alpine_hometown_df = df2.rename(columns={ df2.columns[0]: "Hometown" })
freeski_name_df = df3.rename(columns={ df3.columns[0]: "Name" })
freeski_hometown_df = df4.rename(columns={ df4.columns[0]: "Hometown" })
freestyle_name_df = df5.rename(columns={ df5.columns[0]: "Name" })
freestyle_hometown_df = df6.rename(columns={ df6.columns[0]: "Hometown" })
xc_name_df = df7.rename(columns={ df7.columns[0]: "Name" })
xc_hometown_df = df8.rename(columns={ df8.columns[0]: "Hometown" })
ski_jumping_name_df = df9.rename(columns={ df9.columns[0]: "Name" })
ski_jumping_hometown_df = df10.rename(columns={ df10.columns[0]: "Hometown" })
nordic_combined_name_df = df11.rename(columns={ df11.columns[0]: "Name" })
nordic_combined_hometown_df = df12.rename(columns={ df12.columns[0]: "Hometown" })
snowboard_name_df = df13.rename(columns={ df13.columns[0]: "Name" })
snowboard_hometown_df = df14.rename(columns={ df14.columns[0]: "Hometown" })
```

```python
# Combine Name and Hometown for each team
alpine_df_noteam = pd.concat([alpine_name_df,alpine_hometown_df], axis=1)
freeski_df_noteam = pd.concat([freeski_name_df,freeski_hometown_df], axis=1)
freestyle_df_noteam = pd.concat([freestyle_name_df,freestyle_hometown_df], axis=1)
xc_df_noteam = pd.concat([xc_name_df,xc_hometown_df], axis=1)
ski_jumping_df_noteam = pd.concat([ski_jumping_name_df,ski_jumping_hometown_df], axis=1)
nordic_combined_df_noteam = pd.concat([nordic_combined_name_df,nordic_combined_hometown_df], axis=1)
snowboard_df_noteam = pd.concat([snowboard_name_df,snowboard_hometown_df], axis=1)
```

```python
# Add team column to each team DF
alpine_df_noteam["Team"] = "Alpine"
freeski_df_noteam["Team"] = "Freeski"
freestyle_df_noteam["Team"] = "Freestyle"
xc_df_noteam["Team"] = "Cross Country"
ski_jumping_df_noteam["Team"] = "Ski Jumping"
nordic_combined_df_noteam["Team"] = "Nordic Combined"
snowboard_df_noteam["Team"] = "Snowboard"
```

```python
# Update noteam df to final '(team)_df'
alpine_df = alpine_df_noteam
freeski_df = freeski_df_noteam
freestyle_df = freestyle_df_noteam
xc_df = xc_df_noteam
ski_jumping_df = ski_jumping_df_noteam
nordic_combined_df = nordic_combined_df_noteam
snowboard_df = snowboard_df_noteam
```

```python
# Create final US Winter Olympic Team DF by merging all team DFs
us_winter_olympic_team = pd.concat([alpine_df, freeski_df, freestyle_df, xc_df, ski_jumping_df, nordic_combined_df, snowboard_df], axis=0)
```

```python
# Reset the index
us_winter_olympic_team.reset_index()
```

```python
# Split Hometown into City and State

# Search for end of City by looking for the comma and space
new_us_winter_olympic_team = us_winter_olympic_team["Hometown"].str.split(", ", n = 1, expand = True) 
  
# Make Separate City column from new data frame 
us_winter_olympic_team["City"]= new_us_winter_olympic_team[0] 
  
# Make separate State column from new data frame 
us_winter_olympic_team["State"]= new_us_winter_olympic_team[1] 
  
# Drop Hometown column
us_winter_olympic_team.drop(columns =["Hometown"], inplace = True) 
```
